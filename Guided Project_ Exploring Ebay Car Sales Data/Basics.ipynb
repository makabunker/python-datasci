#### Preliminary investigations show that both the 'seller' and 'offer_type' columns contain idential values for every row except 1, so we can eliminate those columns. The 'price' and 'odometer' columns will need to be convereted from an object to a float after removing the dollar sign and changing the column name to mention the price is in Euros and distance in kms. Each column containing dates will also need to be changed from object to a datetime. 

#### The 'registration_month' and 'power_ps' columns both have minimum values of 0, so those will need to be further investigated to make sure 0 values do not affect the analysis. Also, the 'registration_year' column has a min, max of 1000, 9999, which will need to be examined.

```python
years = autos['registration_year'].value_counts()
print(years.sort_index())
print()
print(autos['registration_month'].value_counts().sort_index())
print()
print(autos['power_ps'].value_counts().sort_index())
```

#### Further investigation reveals a handful of entries for cars registered before 1900 and after 2016 which are both impossible (data was scraped in 2016). Those entries will be removed. Additionally, 5000+ cars list the registration month as 0. 

#### The 'power_ps' column lists a range of values from 0-9999. As the average PS of a car is around 120, with the minimum being 1 or 2 for mopeds and the maximum known PS of a car being around 700, we'll need to remove any values outside of those bounds. It is also adviseable to investigate cars toward the lower or higher ends to see if their reported values make sense. 

```python
# Clean and rename 'price' and 'odometer' columns
autos['price_euro'] = autos['price'].str.replace('$', '').str.replace(',', '').astype(float)
autos['odometer_km'] = autos['odometer'].str.replace(',', '').str.replace('km','').astype(float)
autos = autos.drop(['price', 'odometer'], axis=1)
```

```python
# Clean 'registration_year' and 'registration_month' columns
# and drop bad rows
bad_reg_year = (autos['registration_year'] > 2016) | (autos['registration_year'] < 1900)
autos = autos.loc[~bad_reg_year]

bad_reg_months = autos['registration_month'] == 0
autos = autos.loc[~bad_reg_months]

# Clean 'power_ps' column and drop and rows below 1 or above 700
bad_power = (autos['power_ps'] > 700) | (autos['power_ps'] < 1)
autos = autos.loc[~bad_power]

# Drop 'seller' and 'offer_type' columns
autos = autos.drop(['seller', 'offer_type'], axis=1)
```

```python
# import datetime to convert date columns to dtypes datetime for
# date_crawled, ad_created, last_seen columns
import datetime

# Create function to convert values in col to datetime
def col_to_datetime(col):
    col_dt = pd.to_datetime(col, format='%Y-%m-%d %H:%M:%S')
    return col_dt

autos['date_crawled'] = col_to_datetime(autos['date_crawled'])
autos['ad_created'] = col_to_datetime(autos['ad_created'])
autos['last_seen'] = col_to_datetime(autos['last_seen'])
```

#### A majority of the cars listed are of 10 different brands, with Volkswagen dominating the group. Next we will determine the average prices of each brand of car for the top 10 brands in the dataset

```python
top10_brands = autos['brand'].value_counts().head(10).index
```

```python
# Create dictionary to hold median price values
top10_brands_median_price = {}

# Loop through the top 10 brands and, for each, find the median
# of the 'price_euro' column in autos. Add brand and price to dictionary
for brand in top10_brands:
    median = autos.loc[autos['brand'] == brand, 'price_euro'].median()
    top10_brands_median_price[brand] = median
    
# top10_brands_median_price = sorted(top10_brands_median_price.items(), key=lambda kv:kv[1], reverse=True)
top10_brands_median_price
```

#### As noted before, due to the possible presence of extremely rare/collectors cars, we are using the median price here as a measure of average. 

####  Audi tops the list with an average price of 7.100,00EUR, closely followed by BMW. 

#### Renault cars average about one-quarter of the price of Audis listed at 1.550,00EUR

#### While Volkswagen is by far the most common car brand in the data set, it is, on average, only the 4th most expensive car of the top 10 brands listed, with an average price of 3.549,00EUR.

```python
# Calculate the mean mileage for each car of the top 10 brands
top10_mean_mileage = {}
for brand in top10_brands:
    mean_mileage = autos.loc[autos['brand'] == brand, 'odometer_km'].mean()
    top10_mean_mileage[brand] = mean_mileage

# Convert dictionaries to pd.Series    
bmp_series = pd.Series(top10_brands_median_price)    
bmm_series = pd.Series(top10_mean_mileage)

# Create a pd.DataFrame from first series and then add the other
top10_price_miles = pd.DataFrame(bmm_series, columns=['mean_mileage'])
top10_price_miles['median_price'] = bmp_series
top10_price_miles.round()
```

#### With the average mileage and price next to each other in our dataframe, we can see that there doesn't appear to be a correlation between the two. The range of average mileages is very narrow and such small variation should not be expected to explain the large price differential
